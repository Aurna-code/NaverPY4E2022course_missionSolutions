조건 1: 게임을 몇 판을 진행할 것인지 입력을 받아주기

조건 2: 0, 1, 2, "가위", "바위", "보" 이외에 다른 입력을 받으면 재입력 받기

조건 3: 게임종료 후 나와 컴퓨터의 총 전적 출력하기

Q3. 2개의 숫자를 입력하여 그 사이에 짝수만 출력하는 함수를 만들어 봅시다. 그리고 중앙값도 함께 출력 해봅시다.(단, 중앙값이 짝수가 아닐 경우에는 중앙값은 출력을 하지 않고, 짝수인 수만 출력한다)


Q4. 2개의 숫자를 입력하여 그 사이에 소수가 몇 개인지 출력하는 함수를 만들어 봅시다.

소수: 1과 자기 자신만을 약수로 가지는 수

```python
import random

class third_mission:
  def __init__(self):
    while 1:
      print("3주차 미션용 클래스는 구구단 함수, 가위바위보 업그레이드 버전 함수, 짝수-중앙값 출력 함수, 소수 출력 함수로 이루어져 있습니다.\n'구구단', '가위바위보', '짝수', '소수' 중 하나를 입력해 주십시오.")
      print("클래스를 종료하시려면 '종료'를 입력해 주십시오.")
      sel = input()
      if sel == '구구단':
        self.multi_table()
      elif sel == '가위바위보':
        self.srp_scan()
      elif sel == '짝수':
        self.even_median()
      elif sel == '소수':
        self. decimal()
      elif sel == '종료':
        return
      else:
        continue

  def multi_table(self):
    while 1:
      print("구구단 함수가 실행되었습니다. 찾고자 하는 구구단의 숫자를 입력해 주십시오.\n구구단을 종료하려면 '종료'를 입력하십시오.")
      num = input()
      if num.isdigit(): 
        for i in range(1, 10, 2):
          res = int(num) * i
          if res <= 50:
            sets  = f"{num} * {i} = {res}"
            print(sets)

      elif "단" in num:
        for j in num:
          if j.isdigit():
            for i in range(1, 10, 2):
              res = int(j) * i
              if res <= 50:
                sets  = f"{j} * {i} = {res}"
                print(sets)

      elif num == '종료':
        break
      else:
        print("잘못된 입력입니다.")
  
  def srp_scan(self):
    while 1:
      print("가위바위보 함수가 실행되었습니다. 진행할 게임의 횟수를 입력해 주십시오.\n가위바위보를 종료하려면 '0'을 입력하십시오.")
      turns = input()
      if turns.isdigit():
        if int(turns) != 0 and int(turns) > 0:
          win = []
          lose = []
          draw = 0
          for i in range(int(turns)):
            while 1:
              print("가위, 바위, 보 혹은 0, 1, 2 중 하나를 입력해 주십시오.")
              you = input()
              if you == '가위' or you == '바위' or you == '보' or you == '0' or you == '1' or you == '2':
                self.srp(you)
                if self.result == 0:
                  draw += 1
                  break
                elif self.result == 1 or self.result == -2:
                  win.append(self.result)
                  break
                else:
                  lose.append(self.result)
                  break
              else:
                print("잘못된 입력입니다.")
          report = f"{len(win)}번 승리, {len(lose)}번 패배, {draw}번 비겼습니다." 
          print(report)
        elif int(turns) == 0:
          return
        else:
          continue
      else:
        continue

  def srp(self, you):
    you_table_1 = {"가위":0, "바위":1, "보":2}
    you_table_2 = {0 : "가위", 1 : "바위", 2 : "보"}
    res_table = {0 : "비겼습니다.", 1 : "당신이 이겼습니다.", 2 : "당신이 졌습니다.", -1 : "당신이 졌습니다." , -2 : "당신이 이겼습니다."}

    com = random.randint(0,2)
    for key, value in you_table_1.items():
      if value == com:
        convert = key

    if you.isdigit() == 0:
      winlose = f"당신 : {you}, 컴퓨터 : {convert}"
      self.result = you_table_1[you] - com
    else:
      winlose = f"당신 : {you_table_2[int(you)]}, 컴퓨터 : {convert}"
      self.result = int(you) - com

    print(winlose)
    print(res_table[self.result])
  
  def even_median(self):
    while 1:
      try:
        print("두 값 사이의 짝수와 중앙값을 출력해주는 함수입니다. 낮은 값과 높은 값을 한 칸 띄어서 정수로 입력해 주십시오.")
        value1, value2 = input().split()
        value1 = int(value1)
        value2 = int(value2)
      except ValueError:
        print("정확하지 않은 입력입니다.")
        continue
      if value1 == value2:
        print("낮은 값과 높은 값의 차이가 없습니다.")
      elif value1 > value2:
        high = value1
        low = value2
        break
      else:
        high = value2
        low = value1
        break

    numbers = [i for i in range(low, high+1)]
    evens = [i for i in numbers if i%2 == 0]

    if len(numbers) <= 2:
      median = (numbers[0] + numbers[1]) / 2 
    else:
      if len(numbers) % 2 == 0:
        median = (numbers[int(len(numbers)/2) - 1] + numbers[int(len(numbers)/2)]) / 2
      else:
        median = numbers[int((len(numbers) + 1)/2) - 1]

    if len(evens) >= 1:
      print(f"두 값 사이의 짝수는 {evens}이며, 중앙값은 {median}입니다.\n===============")
    else:
      print(f"두 값 사이에 짝수는 없으며, 중앙값은 {median}입니다.\n===============")

  def decimal(self):
    while 1:
      print("이 함수는 입력된 두 값 사이의 소수를 출력합니다.\n두 값을 입력해 주십시오.")
      val1 = input("값 1 : ")
      val2 = input("값 2 : ")

      if val1.isdigit == False or val2.isdigit == False:
        print("잘못된 입력입니다.")
      else:
        val1 = int(val1)
        val2 = int(val2)
        if val1 == val2:
          print("동일한 값을 입력할 수 없습니다.")
        elif val1 > val2:
          high = val1
          low = val2
          break
        else:
          high = val2
          low = val1
          break
    
    nums = [i for i in range(low+1, high)]
    deci = []
    for i in nums:
      flag = True
      if i > 1:
        for j in range(2, i):
          if i % j == 0:
            flag = False
        if flag:
          deci.append(i)
                      
    if len(deci) < 1:
      print("입력하신 두 값 사이에 소수가 없습니다.\n=========== ")
    else:
      print(f"두 값 사이의 소수는 {deci}입니다.\n============")



third_mission()
```
