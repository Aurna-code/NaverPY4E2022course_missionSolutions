Q3. 숫자 맞추기 게임을 만들어 볼게요. 컴퓨터가 임의의 숫자를 3개 만들고 우리가 그것을 맞춰보겠습니다.

조건1 - 숫자는 0 ~ 100까지 숫자를 3개 만듭니다(중복 불가).

조건2 - 5회, 10회까지 정답을 못맞추면 최솟값, 최대값에 대한 힌트를 줍니다.

조건3 - 정답을 맞추면, 맞춘 정답이 최솟값인지, 중간값인지, 최댓값인지 알려줍니다.

```python
import random

def guess_numbers_generate():
    print("숫자맞추기 게임에 오신 것을 환영합니다. 컴퓨터가 만든 1~100 사이의 숫자 3개를 맞춰보세요.")
    while 1:
      number = random.randint(0, 100)
      number_2 = random.randint(0, 100)
      number_3 = random.randint(0, 100)
      if number != number_2 and number != number_3 and number_2 != number_3:
        break
   
    number_list = [number, number_2, number_3]
    high = max(number_list)
    low = min(number_list)
    for i in number_list:
      if i != high and i != low:
        med = i
    guess_numbers(high, low, med)
    
def guess_numbers(high, low, med):
   count = 0
   my_numberlist = []
   low_check = False
   high_check = False
   med_check = False
   while 1:
    my_number = input("숫자를 예측해보세요 : ")
    my_number = int(my_number)
    count +=1
    if my_number in my_numberlist:
      print(f"{count}차 시도\n이미 예측에 사용한 숫자입니다")
      continue
    elif my_number == low:
      print(f"{count}차 시도\n숫자를 맞추셨습니다! {my_number}은(는) 최솟값입니다.")
      low_check = True
    elif my_number == high:
      print(f"{count}차 시도\n숫자를 맞추셨습니다! {my_number}은(는) 최대값입니다.")
      high_check = True
    elif my_number == med:
      print(f"{count}차 시도\n숫자를 맞추셨습니다! {my_number}은(는) 중간값입니다.")
      med_check = True
    else:
      print(f"{count}차 시도\n{my_number}은(는) 없습니다")
      if count >=5:
        guess_numbers_hint(count, my_number, low, med, high)

    if low_check == True and high_check == True and med_check == True:
      print(f"게임종료\n{count}시도만에 예측 성공")
      return
    
    my_numberlist.append(my_number)
      
    

def guess_numbers_hint(count, my_number, low, med, high):
  if my_number > high:
    if count >= 10:
      if my_number - high >= 10:
        print(f"{my_number}은(는) 최대값보다 10 이상의 차이로 큽니다")
        return
      else:
        print(f"{my_number}은(는) 최대값보다 10 미만의 차이로 큽니다")
        return
    else:
      print(f"최대값은 {my_number}보다 작습니다")
      return
  elif my_number > med:
    if count >= 10:
      print(f"{my_number}은(는) 최대값과 중앙값 사이에 있습니다")
      return
    else:
      print(f"중앙값은 {my_number}보다 작습니다")
      return
  elif my_number > low:
    if count >= 10:
      print(f"{my_number}은(는) 중앙값과 최솟값 사이에 있습니다.")
      return
    else:
      print(f"최솟값은 {my_number}보다 작습니다")
      return
  else:
    if count >= 10:
      if low - my_number >= 10:
        print(f"{my_number}은(는) 최솟값보다 10 이상의 차이로 작습니다")
      else:
        print(f"{my_number}은(는) 최솟값보다 10 미만의 차이로 작습니다")
    else:
      print(f"최솟값은 {my_number}보다 큽니다")
      return

guess_numbers_generate()

  
```
