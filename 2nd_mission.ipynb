          (0, 1 ,2 혹은 "가위", "바위", "보" 로 입력할 수 있습니다. - 총 6가지 방법으로 넣을 수 있음)

Q3. 학생 이름과 점수를 입력하면 학점을 출력하는 학점 변환기를 만들어 봅시다.

```python
import random
import pandas as pd

class second_mission:
  def __init__(self):
    while 1:
      print("2주차 미션용 클래스 코드는 4가지의 함수를 포함하고 있습니다.")
      print("가위바위보 게임을 실행하려면 '가위바위보', 연봉 계산기를 실행하려면 '연봉',\n학점 계산기를 실행하려면 '학점',버스 요금 계산기를 실행하려면 '버스'를 입력해 주십시오.\n클래스를 종료하시려면 '종료'를 입력하십시오.")
      self.scan = input()
      if self.scan == '가위바위보' or self.scan == '연봉' or self.scan == "학점" or self.scan == "버스":
        self.select()
      elif self.scan == '종료':
        break
      else:
        continue

  def select(self):
    while 1:
      if self.scan == "가위바위보":
        self.SRP()
        break
      elif self.scan == "연봉":
        self.salary()
        break
      elif self.scan == "학점":
        self.credit_input()
        break
      elif self.scan == "버스":
        self.bus_input()
        break
      else: 
        print("잘못된 명령어입니다.")
        break

  def SRP(self):   
    print("가위바위보 게임을 시작합니다.")
    while 1 :
      print("가위, 바위, 보 중 하나를 입력하거나 0, 1, 2의 숫자를 입력해 주십시오.\n게임을 종료하시려면 아무 값이나 입력해 주십시오.")
      my = input()
      if my == "가위" or my == "바위" or my == "보":
        self.rcp(my)
      elif my == "0" or my == "1" or my == "2":
        self.rcp(my)
      else:
        return

  def rcp(self, my):
    while 1:
      com = random.randint(0,2)
      if my == "가위" or 0:
        if com == 0:
          print("당신 : 가위, 컴퓨터 : 가위")
          print("비겼습니다! \n게임을 다시 하려면 r키를 입력해 주십시오.")
          scann = input()
          if scann == 'r':
            self.SRP()
          else:
            return

        elif com == 1:
          print("당신 : 가위, 컴퓨터 : 바위")
          print("컴퓨터 승리!.\n게임을 다시 하려면 r키를 입력해 주십시오.")
          scann = input()
          if scann == 'r':
            self.SRP()
          else:
            return
        elif com == 2:
          print("당신 : 가위, 컴퓨터 : 보")
          print("승리!\n게임을 다시 하려면 r키를 입력해 주십시오.")
          scann = input()
          if scann == 'r':
            self.SRP()
          else:
            break
      if my == "바위" or 1:
        if com == 0:
          print("당신 : 바위, 컴퓨터 : 가위")
          print("승리!\n게임을 다시 하려면 r키를 입력해 주십시오.")
          scann = input()
          if scann == 'r':
            self.SRP()
          else:
            return
        elif com == 1:
          print("당신 : 바위, 컴퓨터 : 바위")
          print("비겼습니다.\n게임을 다시 하려면 r키를 입력해 주십시오.")
          scann = input()
          if scann == 'r':
            self.SRP()
          else:
            return
        elif com == 2:
          print("당신 : 바위, 컴퓨터 : 보")
          print("컴퓨터 승리!\n게임을 다시 하려면 r키를 입력해 주십시오.")
          scann = input()
          if scann == 'r':
            self.SRP()
          else:
            return
      if my == "보" or 2:
        if com == 0:
          print("당신 : 보, 컴퓨터 : 가위")
          print("컴퓨터 승리!\n게임을 다시 하려면 r키를 입력해 주십시오.")
          scann = input()
          if scann == 'r':
            self.SRP()
          else:
            return
        elif com == 1:
          print("당신 : 보, 컴퓨터 : 바위")
          print("승리!\n게임을 다시 하려면 r키를 입력해 주십시오.")
          scann = input()
          if scann == 'r':
            self.SRP()
          else:
            return
        elif com == 2:
          print("당신 : 보, 컴퓨터 : 보")
          print("비겼습니다.\n게임을 다시 하려면 r키를 입력해 주십시오.")
          scann = input()
          if scann == 'r':
            self.SRP()
          else:
            return
      else:
        return
      
  def salary(self):
    print("연봉 계산기를 실행합니다. 이 계산기는 세전 연봉과 세후 연봉을 동시에 출력합니다.\n월급을 입력해 주십시오.")
    monthly_payment = input()
    monthpay = []
    for i in monthly_payment:
      if i.isdigit():
        if int(i) < 0:
          print("0보다 작은 수는 입력할 수 없습니다.")
          break
        else:
          monthpay.append(i)
    monthpay_attach = [monthpay[0]]
    index = 1
    while index < len(monthpay):
      monthpay_attach[0] = monthpay_attach[0] + monthpay[index]
      index += 1
    monthpay_attach[0] = int(monthpay_attach[0])

    checklist_1 = ['만원', '만 원', '만']
    checklist_2 = ['억원', '억 원', '억']

    for i in checklist_1:
      if i in monthly_payment:
        monthpay_attach[0] = monthpay_attach[0] * 120000
        break
    for i in checklist_2:
      if i in monthly_payment:
        monthpay_attach[0] = monthpay_attach[0] * 1200000000
        break

    self.yearly_payment(monthpay_attach)


  def yearly_payment(self, monthpay_attach):
    frame = {'1200만원 이하' : ['6%'],
             '4600만원 이하' : ['15%'],
             '8800만원 이하' : ['24%'],
             '1억 5천만원 이하' : ['35%'],
             '3억원 이하' : ['38%'],
             '5억원 이하' : ['40%'],
             '5억원 초과' : ['42%']}
    rules = pd.DataFrame(frame)
    
    monthly_payment = monthpay_attach[0]
    criteria_1 = int(rules.columns[0].split('만')[0]) * 10000
    criteria_2 = int(rules.columns[1].split('만')[0]) * 10000
    criteria_3 = int(rules.columns[2].split('만')[0]) * 10000
    criteria_4 = int(rules.columns[3].rsplit(' ', 1)[0].split('억')[0] + rules.columns[3].rsplit(' ', 1)[0].split('억')[1].split('천만')[0].replace(" ", "")) * 10000000
    criteria_5 = int(rules.columns[4].split('억')[0]) * 100000000
    criteria_6 = int(rules.columns[5].split('억')[0]) * 100000000
    criteria_7 = int(rules.columns[5].split('억')[0]) * 100000000


    if monthly_payment <= criteria_1:
      after_payment = monthly_payment * ((100 - int(rules.iloc[0,0].split("%")[0])) / 100)
    elif monthly_payment <= criteria_2:
      after_payment = monthly_payment * ((100 - int(rules.iloc[0,1].replace("%", "")) / 100))
    elif monthly_payment <= criteria_3:
      after_payment = monthly_payment * ((100 - int(rules.iloc[0,2].split("%")[0])) / 100)
    elif monthly_payment <= criteria_4:
      after_payment = monthly_payment * ((100 - int(rules.iloc[0,3].split("%")[0])) / 100)
    elif monthly_payment <= criteria_5:
      after_payment = monthly_payment * ((100 - int(rules.iloc[0,4].split("%")[0])) / 100)
    elif monthly_payment <= criteria_6:
      after_payment = monthly_payment * ((100 - int(rules.iloc[0,5].split("%")[0])) / 100)
    else:
      after_payment = monthly_payment * ((100 - int(rules.iloc[0,6].split("%")[0])) / 100)

    print("세전 연봉 : {}원".format(monthly_payment))
    print("세후 연봉 : {}원".format(int(after_payment)))
    print("---------------------")


  def credit_input(self):
    print("성적 계산기는 학생의 성적을 바탕으로 등급을 매겨 드립니다.\n학생의 이름을 입력해 주십시오.")
    name = input()
    print("학생의 점수를 입력해 주십시오.")
    while 1:
      score = input()
      if int(score) < 0 or int(score) > 100:
        print("100점 이하 0점 이상의 점수를 입력해 주십시오.")
      else:
        break
    self.grader(name, score)

  def grader(self, name, score):
    print("학생이름 : {}".format(name))
    print("점수 : {}점".format(score))
    score = int(score)
    while 1:
      if score >=95 and score <= 100:
        grade = "A+"
        break
      elif score >= 90:
        grade = "A"
        break
      elif score >= 85:
        grade = "B+"
        break
      elif score >= 80:
        grade = "B"
        break
      elif score >= 75:
        grade = "C+"
        break
      elif score >= 70:
        grade = "C"
        break
      elif score >= 65:
        grade = "D+"
        break
      elif score >= 60:
        grade = "D"
        break
      elif score >= 0:
        grade = "F"
        break
      else:
        print("잘못된 입력입니다.")
        continue
    print("학점 : {}".format(grade))
    print(" -------------- ")

  def bus_input(self):
    print("버스 요금 계산기는 연령과 요금 지불수단을 바탕으로 버스 요금을 계산해 드립니다.")
    print("연령을 입력해 주십시오.")
    while 1:
      age = input()
      if int(age) < 0:
        print("0 이상의 숫자를 입력해 주십시오.")
      else:
        break
    print("'카드' 혹은 '현금' 중에서 요금 지불수단을 입력해 주십시오.")
    while 1:
      pay = input()
      if pay is '카드' or pay is '현금':
        print("'카드' 혹은 '현금'을 입력해 주십시오.")
      else:
        break
    self.bus_fare(age, pay)
  
  def bus_fare(self, age, pay):
    print("나이 : {}세".format(age))
    print("지불유형 : {}".format(pay))
    age = int(age)
    if age < 8:
      fare = "무료"
    elif age >= 8 and age < 14:
      fare = "450원"
    elif age >= 14 and age <20:
      if pay == "카드":
        fare = "720원"
      else:
        fare = "1000원"
    elif age >= 20 and age < 75:
      if pay == "카드":
        fare = "1200원"
      else:
        fare = "1300원"
    else:
      fare = "무료"
    print("버스요금 : {}".format(fare))
    print("----------------")
   

s = second_mission()




```
